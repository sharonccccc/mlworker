````python
from autogluon.tabular import TabularDataset, TabularPredictor
data_url = 'https://raw.githubusercontent.com/mli/ag-docs/main/knot_theory/'
train_data = TabularDataset(f'{data_url}train.csv')
train_data.head()
````

````python
save_path = "./model/"
label = 'signature'
predictor = TabularPredictor(label="signature", path=save_path)
````

````python
# from autogluon.tabular import TabularDataset, TabularPredictor
# train_data = TabularDataset('diamonds.csv')
# save_path = "./model/"
# predictor = TabularPredictor(label="Price", path=save_path)
````

````python
predictor.fit(
    train_data=train_data,
    ag_args_fit={
        'num_cpus': 4, 
        'num_gpus': 1
    },  # 啟用 GPU
    # 可以限制時間以快速驗證
    time_limit=300  # 5分鐘
)
````

````python
predictor.leaderboard(display=True)
````

````python
aaa = TabularPredictor.load(save_path)
````

````python
test_sample = train_data.head(5)
test_sample
````

````python
y_pred = aaa.predict(test_sample)
y_pred
````
